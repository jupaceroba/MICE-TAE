# Regresión Lineal

## Introducción 

El análisis de regresión es una de las técnicas de uso más usado para el análisis de datos con numerosas variables. Su atractivo y utilidad general son el resultado del proceso conceptualmente lógico de usar una ecuación para expresar la relación entre una variable de interés (la respuesta o dependiente) y un conjunto de variables predictoras relacionadas (o también llamadas varibles independientes o explicativas). El análisis de regresión busca, por medio de una teoría matemática de fondo, abstraer un problema de la realidad y plasmarlo en lo que llamaremos un modelo.

Esta teoría estadística de fondo de la que hablamos Se ha manisfestado en diferentes metodologías, de las cuales resaltamos el método de mínimos cuadrados y el método de máxima verosimilítud. ¿Pero en qué radica la importancia de estos? bueno, estos nos ayudan a encontrar lo que llamaremos los parámetros del modelo, los cuales son los que nos ayudan a dar respuesta al análisis del fenómeno que está detrás de esto que denominamos regresión. Pero, qué ocurre cuando nos enfrentamos a una cantidad de datos tan númerosos? bueno, aquí el problema radica en la perdida de tiempo que implica realizar el cálculo matemático de estos paramétros a mano. Ante esto, es de mucha importancia el apoyo computacional, que nos permita hacer estos cálculos de forma muy precisa y eficaz y así economizar tiempo. Algunos ejemplos de software que nos permiten hacer estos análisis son R, Python, SAS, entre otros.

El trabajo presentado en este documento abordará alguna de las metodologías aquí mencionadas aplicada a un conjunto de datos y con ayuda del sotware Python se dará respuesta al fénomeno que está detrás del modelo de regresión ajustado a estos datos.$^1$

## Breve historia

Las nociones de la regesión lineal han tenido diferentes bases a lo largo de la historia proveninentes de diferentes autores. Los primeros que empezaron a incursionar con esto fueron Gauss y legendre con el concepto de los mínimos cuadrados para ajustar los datos sobre las órbitas de cuerpos celestes. Fue en 1805 cuando Legendre hizo la primera publicación con respecto al método de mínimos cuadrados ordinarios, en la que el muestra que esta técnica de optimización busca encontrar la función
(dentro de una familia) que mejor se ajusta a los datos bivariantes, de acuerdo con el criterio de mínimo error cuadrático. El problema que tenía Legendre con esto era que a pesar de que se conocían las ecuaciones funcionales de estas órbitas, los errores de medida hacía que los cálculos fuesen aproximados, generando sesgos en las estimaciones de estos estimadores de mínimos cuadrados. En 1829 Gauss fue capaz de establecer la razón por la cual este procedimiento es muy adecuado desde el punto de vista estadístico, mediante lo que hoy se conoce como teorema de Gauss-Markov, que muestra que los estimadores obtenidos con este método son insesgados.$^2$

## Planteamiento del algoritmo

El modelo de regresión múltiple
EI modelo que se plantea en regresión mültiple es el siguiente:

$y_{i}=\beta_{0}+\beta_{1} x_{1 i}+\beta_{2} x_{2 i}+\ldots+\beta_{k} x_{k i}+\epsilon_{i}$

Donde, $x_{1}, x_{2}, \ldots, x_{k}$ son las variables independientes o explicativas.
componente de êror que se distribuye según una normal: $u_{i}=N\left(0, \sigma^{2}\right)$
El ajuste del modelo se realiza por el método de máxima verosimilitud oel método de mínimos cuadrados. En el caso de distribución normal de errores, ambos métodos coinciden, como ya se vió en regresión simple.
O de forma matricial esto se podría escribir como se muestra a continuación:

$$
\mathbf{y}=\mathbf{X} \beta+\varepsilon
$$
en donde
$$
\begin{array}{l}
\mathbf{y}=\left[\begin{array}{c}
y_{1} \\
y_{2} \\
\vdots \\
y_{n}
\end{array}\right], & \mathbf{X}=\left[\begin{array}{ccccc}
1 & x_{11} & x_{12} & \cdots & x_{1 k} \\
1 & x_{21} & x_{22} & \cdots & x_{2 k} \\
\vdots & \vdots & \vdots & & \vdots \\
1 & x_{n 1} & x_{n 2} & \cdots & x_{n k}
\end{array}\right] \\
\beta & =\left[\begin{array}{c}
\beta_{0} \\
\beta_{1} \\
\vdots \\
\beta_{k}
\end{array}\right], & \varepsilon=\left[\begin{array}{c}
\varepsilon_{1} \\
\varepsilon_{2} \\
\vdots \\
\varepsilon_{n}
\end{array}\right]
\end{array}
$$

Con respecto a lo que tenemos atrás se desea encontrar los coeficientes del modelo mediante el método de mínimos cuadrados tal y como se muestra a continuación:
$$
S(\beta)=\sum_{i=1}^{n} \varepsilon_{i}^{2}=\varepsilon^{\prime} \varepsilon=(\mathbf{y}-\mathbf{X} \beta)^{\prime}(\mathbf{y}-\mathbf{X} \boldsymbol{\beta})
$$
Nótese que $S(\beta)$ se puede expresar como sigue:
$$
\begin{aligned}
S(\beta) &=\mathbf{y}^{\prime} \mathbf{y}-\beta^{\prime} \mathbf{X}^{\prime} \mathbf{y}-\mathbf{y}^{\prime} \mathbf{X} \beta+\beta^{\prime} \mathbf{X}^{\prime} \mathbf{X} \beta \\
&=\mathbf{y}^{\prime} \mathbf{y}-2 \beta^{\prime} \mathbf{X}^{\prime} \mathbf{y}+\beta^{\prime} \mathbf{X}^{\prime} \mathbf{X} \beta
\end{aligned}
$$
ya que $\beta^{\prime} \mathbf{X}^{\prime} \mathbf{y}$ es una matriz de $1 \times 1,$ es decir, un escalar, y que su transpuesta $\left(\beta^{\prime} \mathbf{X}^{\prime} \mathbf{y}\right)^{\prime}=$ $\mathbf{y}^{\prime} \mathbf{X} \beta$ es el mismo escalar. Los estimadores de mínimos cuadrados deben satisfacer
$$
\left.\frac{\partial S}{\partial \beta}\right|_{\hat{\beta}}=-2 \mathbf{X}^{\prime} \mathbf{y}+2 \mathbf{X}^{\prime} \mathbf{X} \hat{\beta}=\mathbf{0}
$$
que se simplifica a
$$
\mathbf{X}^{\prime} \mathbf{X} \hat{\beta}=\mathbf{X}^{\prime} \mathbf{y}

Las ecuaciones anteriores son las ecuaciones normales de mínimos cuadrados. Son la forma matricial de la presentación escalar, ecuaciones.

Para resolver las ecuaciones normales se multiplican ambos lados de por la inversa de $\mathbf{X}^{\prime} \mathbf{X}$. Así, el estimador de $\beta$ por mínimos cuadrados es
$$
\hat{\beta}=\left(\mathbf{X}^{\prime} \mathbf{X}\right)^{-1} \mathbf{X}^{\prime} \mathbf{y}
$$
siempre y cuando exista la matriz inversa $\left(\mathbf{X}^{\prime} \mathbf{X}\right)^{-1}$. La matriz $\left(\mathbf{X}^{\prime} \mathbf{X}\right)^{-1}$ siempre existe si los regresores son linealmente independientes, esto es, si ninguna columna de la matriz $\mathbf{X}$ es una combinación lineal de las demás columnas.

Es fácil de ver que la forma matricial de las ecuaciones normales es idéntica a la forma escalar. Al escribir con detalle se obtiene

$$
\left[\begin{array}{ccccc}
n & \sum_{i=1}^{n} x_{i 1} & \sum_{i=1}^{n} x_{i 2} & \cdots & \sum_{i=1}^{n} x_{i k} \\
\sum_{i=1}^{n} x_{i 1} & \sum_{i=1}^{n} x_{i 1}^{2} & \sum_{i=1}^{n} x_{i 1} x_{i 2} & \cdots & \sum_{i=1}^{n} x_{i 1} x_{i k} \\
\vdots & \vdots & \vdots & & \vdots \\
\sum_{i=1}^{n} x_{i k} & \sum_{i=1}^{n} x_{i k} x_{i 1} & \sum_{i=1}^{n} x_{i k} x_{i 2} & \cdots & \sum_{i=1}^{n} x_{i k}^{2}
\end{array}\right]\left[\begin{array}{c}
\hat{\beta}_{0} \\
\hat{\beta}_{1} \\
\vdots \\
\hat{\beta}_{k}
\end{array}\right]=\left[\begin{array}{c}
\sum_{i=1}^{n} y_{i} \\
\sum_{i=1}^{n} x_{i 1} y_{i} \\
\vdots \\
\sum_{i=1}^{n} x_{i k} y_{i}
\end{array}\right]
$$
Si se hace la multiplicación matricial indicada, se obtiene la forma escalar de las ecuaciones normales (3.11). En esta presentación se ve que $\mathbf{X}^{\prime} \mathbf{X}$ es una matriz simétrica de $p \times p, y$ que $\mathbf{X}^{\prime} \mathbf{y}$ es un vector columna de $p \times 1$. Nótese la estructura especial de la matriz $\mathbf{X}^{\prime} \mathbf{X}$. Los elementos diagonales de $\mathbf{X}^{\prime} \mathbf{X}$ son las sumas de los cuadrados de los elementos en las columnas de $\mathbf{X}$, y los elementos fuera de la diagonal son las sumas de los productos cruzados de los elementos de las columnas de $\mathbf{X}$. Además, nótese que los elementos de $\mathbf{X}^{\prime} \mathbf{y}$ son las sumas de los productos cruzados de las columnas de $\mathbf{X}$ por las observaciones $y_{i}$

El modelo ajustado de regresión que corresponde a los niveles de las variables regresoras $\mathbf{x}^{\prime}=\left[1, x_{1}, x_{2}, \ldots, x_{k}\right]$ es
$$
\hat{y}=\mathbf{x}^{\prime} \hat{\beta}=\hat{\beta}_{0}+\sum_{j=1}^{k} \hat{\beta}_{j} x_{j}
$$
El vector de valores ajustados $\hat{y}_{i}$ que corresponden a los valores observados $y_{i}$ es
$$
\hat{y}=\mathbf{X} \hat{\beta}=\mathbf{X}\left(\mathbf{X}^{\prime} \mathbf{X}\right)^{-1} \mathbf{X}^{\prime} \mathbf{y}=\mathbf{H} \mathbf{y}
$$
La $n \times n$ es la matriz $\mathbf{H}=\mathbf{X}\left(\mathbf{X}^{\prime} \mathbf{X}\right)^{-1} \mathbf{X}^{\prime}$ se suele llamar matriz de sombrero. Aplica el vector de valores observados en un vector de valores ajustados. La matriz de sombrero y sus propiedades desempeñan un papel central en el análisis de regresión.

La diferencia entre el valor observado $y_{i} \mathrm{y}$ el valor ajustado $\hat{y}_{i}$ correspondiente es el residual $e_{i}=y_{i}-\hat{y}_{i} .$ Los $n$ residuales se pueden escribir cómodamente con notación matricial como sigue:
$$
\mathbf{e}=\mathbf{y}-\hat{y}
$$
Hay otras maneras de expresar el vector de residuales e, que pueden ser útiles, como
$$
\mathbf{e}=\mathbf{y}-\mathbf{X} \hat{\beta}=\mathbf{y}-\mathbf{H} \mathbf{y}=(\mathbf{I}-\mathbf{H}) \mathbf{y}
$$
Teniendo en cuenta la teoría de mostrada atrás se usará statmodels para poder ajustar un modelo de regresión lineal al conjunto de datos que se muestra a continuación.

### Resultado del modelo

```python
data=sm.tools.tools.add_constant(data, prepend=True, has_constant='skip')
fit = sm.OLS(data["Peso"], data[["const", "Sexo", "Años", "Tamaño"]]).fit()
print(fit.summary())
```
